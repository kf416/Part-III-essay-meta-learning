## plot for hyper posterior particles

```python
import matplotlib.pyplot as plt

# Data
x = [1, 3, 5, 10]
avg_rmse = [0.2097, 0.2071, 0.1991, 0.2015]
rmse_std = [0.0205, 0.0194, 0.0168, 0.0171]

cal_err = [0.0572, 0.0489, 0.0451, 0.0485]
cal_std = [0.0348, 0.0248, 0.0182, 0.0194]

avg_ll = [0.0971, 0.1642, 0.1815, 0.1692]
ll_std = [0.1237, 0.0847, 0.0826, 0.0673]

# Plot 1 - Average RMSE
plt.plot(x, avg_rmse, '-o', label='RMSE', color='blue')
plt.fill_between(x, [avg_rmse[i]-rmse_std[i] for i in range(len(x))], 
                [avg_rmse[i]+rmse_std[i] for i in range(len(x))], alpha=0.2, color='blue')
#plt.title('Average RMSE')
plt.xlabel('Number of hyper-posterior particles')
plt.ylabel('Average RMSE')
#plt.legend()
plt.show()

# Plot 2 - Calibration Error
plt.plot(x, cal_err, '-o', label='Calibration Error', color='red')
plt.fill_between(x, [cal_err[i]-cal_std[i] for i in range(len(x))], 
                [cal_err[i]+cal_std[i] for i in range(len(x))], alpha=0.2, color='red')
#plt.title('Calibration Error')
plt.xlabel('Number of hyper-posterior particles')
plt.ylabel('Calibration Error')
#plt.legend()
plt.show()
```

## plot for number of meta train iterations

```python
import matplotlib.pyplot as plt

# Data
iterations = [1000, 5000, 10000, 20000, 50000]

avg_rmse = [0.2630, 0.2417, 0.2057, 0.2071, 0.2078]
std_rmse = [0.0503, 0.0375, 0.0169, 0.0194, 0.0212]

avg_cal_err = [0.1289, 0.1144, 0.0610, 0.0489, 0.0473]
std_cal_err = [0.0383, 0.0336, 0.0245, 0.0248, 0.0241]

# RMSE plot
fig, ax = plt.subplots()

#ax.set_title("RMSE vs. Iteration")
ax.set_xlabel("Meta-Train Iterations")
ax.set_ylabel("Average RMSE")

ax.plot(iterations, avg_rmse,'-o', color='blue')
ax.fill_between(iterations, [avg_rmse[i]-std_rmse[i] for i in range(len(iterations))], 
                [avg_rmse[i]+std_rmse[i] for i in range(len(iterations))], alpha=0.2, color='blue')

plt.show()

# Calibration Error plot
fig, ax = plt.subplots()

#ax.set_title("Calibration Error vs. Iteration")
ax.set_xlabel("Meta-Train Iterations")
ax.set_ylabel("Calibration Error")

ax.plot(iterations, avg_cal_err,'-o', color='red')
ax.fill_between(iterations, [avg_cal_err[i]-std_cal_err[i] for i in range(len(iterations))], 
                [avg_cal_err[i]+std_cal_err[i] for i in range(len(iterations))], alpha=0.2, color='red')

plt.show()
```

## plot for number of context samples

```python
import matplotlib.pyplot as plt

# data for RMSE
x = [1, 2, 5, 10, 15, 20, 30, 40, 50, 60, 80, 100]
avg_rmse = [0.5101, 0.2994, 0.2654, 0.2123, 0.2246, 0.2071, 0.1998, 0.1981, 0.1857, 0.1777, 0.1785, 0.1537]
std_rmse = [0.1292, 0.0366, 0.0253, 0.0177, 0.0228, 0.0194, 0.0196, 0.0177, 0.0171, 0.0220, 0.0156, 0.0147]

# plot RMSE with std deviation
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, avg_rmse, '-o', label='avg_rmse', color='blue')
ax.fill_between(x, [avg_rmse[i]-std_rmse[i] for i in range(len(x))], 
                [avg_rmse[i]+std_rmse[i] for i in range(len(x))], alpha=0.2, color='blue')
ax.set_xlabel('Number of context samples')
ax.set_ylabel('Average RMSE')
ax.legend()
plt.show()
```
